```python
p = 563 # простое число
g = randGen('g', p) # целое число
x = randGen('x', p) # 1 < x < p
```

```python
y = g ** x % p
```

```python
print(f"(p, g, y) = ({p}, {g}, {y}) \n(x) = {x}")
```

```python
#Encryption
```

```python
k = randGen('k', p) # 1 < k < p-1
```

```python
print(f"(k) = {k}")
```

```python
ab_list = []
```

```python
a = g ** k % p
```

```python
print(a)
```

```python
for m in m_list:
    b = (y ** k * m) % p
    ab_list.append((a, b))
```

```python
print(ab_list)
```

```python
writeF('e', ab_list)
```

```python
#Decryption
```

```python
md_list = []
```

```python
for ab in ab_list:
    md = (ab[1] * a ** (p - 1 - x)) % p
    md_list.append(md)
```

```python
print(md_list==m_list)
```

```python
decryption_text = ''.join(chr(i) for i in md_list)
print(decryption_text)
```

```python
writeF('d', decryption_text)
```
